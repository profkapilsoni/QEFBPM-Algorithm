```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
import array as arr
import mitiq
```

```python
from qiskit import IBMQ
IBMQ.save_account('Enter Your Token Here')
IBMQ.load_account()
```

```python
#Text of length 8
qr = QuantumRegister(6)
cr = ClassicalRegister(6)
circuit = QuantumCircuit(qr, cr)

circuit.h(qr[0])
circuit.h(qr[1])
circuit.h(qr[2])

circuit.cx(qr[0], qr[3])

circuit.rccx(qr[1], qr[2], qr[4])

circuit.cx(qr[2], qr[5])
circuit.cx(qr[1], qr[5])
circuit.rccx(qr[1], qr[2], qr[5])

circuit.h(qr[0])
circuit.h(qr[1])
circuit.h(qr[2])

circuit.measure(qr[3], cr[3])
circuit.measure(qr[4], cr[4])
circuit.measure(qr[5], cr[5])
```

```python
provider = IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_qasm_simulator')
job = execute(circuit, backend = qcomp)
from qiskit.tools.monitor import job_monitor
job_monitor(job)
counts = job.result().get_counts()
```

```python
print(counts)
```

```python
plot_histogram(counts)
```

```python
circuit.draw(output = 'mpl')
```

```python
#Text of length 16
qr = QuantumRegister(8)
cr = ClassicalRegister(8)
circuit = QuantumCircuit(qr, cr)

circuit.h(qr[0])
circuit.h(qr[1])
circuit.h(qr[2])
circuit.h(qr[3])

circuit.cx(qr[0], qr[4])

circuit.cx(qr[1], qr[5])

circuit.rccx(qr[2], qr[3], qr[6])
circuit.rccx(qr[0], qr[2], qr[6])
circuit.rcccx(qr[0], qr[2], qr[3], qr[6])

circuit.cx(qr[2], qr[7])
circuit.cx(qr[3], qr[7])
circuit.rccx(qr[0], qr[2], qr[7])
circuit.rccx(qr[2], qr[3], qr[7])
circuit.rcccx(qr[0], qr[2], qr[3], qr[7])
circuit.mcx([qr[0], qr[1], qr[2], qr[3]], qr[7])

circuit.h(qr[0])
circuit.h(qr[1])
circuit.h(qr[2])
circuit.h(qr[3])

circuit.measure(qr[4], cr[4])
circuit.measure(qr[5], cr[5])
circuit.measure(qr[6], cr[6])
circuit.measure(qr[7], cr[7])
```

```python
provider = IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_qasm_simulator')
job = execute(circuit, backend = qcomp)
from qiskit.tools.monitor import job_monitor
job_monitor(job)
counts = job.result().get_counts()
```

```python
print(counts)
```

```python
plot_histogram(counts)
```

```python
circuit.draw(output = 'mpl')
```
