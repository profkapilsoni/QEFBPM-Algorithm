```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
import array as arr
import mitiq
```

```python
from qiskit import IBMQ
IBMQ.save_account('Enter Your Token Here')
IBMQ.load_account()
```

```python
#Text of length 4
qr = QuantumRegister(4)
cr = ClassicalRegister(4)
circuit = QuantumCircuit(qr, cr)

circuit.h(qr[0])
circuit.h(qr[1])
# Quantum state after applying the U_Tpos and U_Qadd unitary 
circuit.cx(qr[1], qr[3]) #f0 0x1
circuit.cx(qr[0], qr[3]) #f0 0x2

circuit.h(qr[0])
circuit.h(qr[1])

circuit.measure(qr[2], cr[2])
circuit.measure(qr[3], cr[3])
```

```python
provider = IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_santiago')
job = execute(circuit, backend = qcomp)
from qiskit.tools.monitor import job_monitor
job_monitor(job)
counts = job.result().get_counts()
```

```python
print(counts)
```

```python
plot_histogram(counts)
```

```python
circuit.draw(output = 'mpl')
```

```python
#Text of length 16
qr = QuantumRegister(8)
cr = ClassicalRegister(8)
circuit = QuantumCircuit(qr, cr)

circuit.h(qr[0])
circuit.h(qr[1])
circuit.h(qr[2])
circuit.h(qr[3])

circuit.cx(qr[3], qr[7]) #f0 0x1
circuit.cx(qr[2], qr[7]) #f0 0x2
circuit.rccx(qr[0], qr[3], qr[7]) #f0 0x9
circuit.rccx(qr[0], qr[2], qr[7]) #f0 0xA

circuit.rcccx(qr[0], qr[1], qr[2], qr[6]) #f1 0xE

circuit.rccx(qr[1], qr[3], qr[5]) #f2 0x5
circuit.rccx(qr[1], qr[2], qr[5]) #f2 0x6
circuit.rcccx(qr[0], qr[1], qr[3], qr[5]) #f2 0xD

circuit.rcccx(qr[0], qr[1], qr[2], qr[4]) #f3 0xE

circuit.h(qr[0])
circuit.h(qr[1])
circuit.h(qr[2])
circuit.h(qr[3])

circuit.measure(qr[4], cr[4])
circuit.measure(qr[5], cr[5])
circuit.measure(qr[6], cr[6])
circuit.measure(qr[7], cr[7])
```

```python
provider = IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_qasm_simulator')
job = execute(circuit, backend = qcomp)
from qiskit.tools.monitor import job_monitor
job_monitor(job)
counts = job.result().get_counts()
```

```python
print(counts)
```

```python
plot_histogram(counts)
```
